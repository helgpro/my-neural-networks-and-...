```python
# Импортируем библиотеки
from sklearn import datasets
import matplotlib.pyplot as plt
 
# Загружаем набор данных
iris_df = datasets.load_iris()
 
# Методы, доступные для набора данных
print(dir(iris_df))
#print(iris_df.data[0:, 1:3 :])# 2 и 3 столбуц

print(iris_df.data[0:,1])# Второй столбец
```

```python
 
# Признаки
print(iris_df.feature_names)
 
# Метки
print(iris_df.target)
 
# Имена меток
print(iris_df.target_names)
 
# Разделение набора данных
x_axis = iris_df.data[:, 0]  # Sepal Length Первый столбец
y_axis = iris_df.data[:, 1]  # Sepal Width второй столбец
 
# Построение
plt.xlabel(iris_df.feature_names[0])
plt.ylabel(iris_df.feature_names[1])
plt.scatter(x_axis, y_axis, c=iris_df.target)
plt.show()
```

### K-means

```python
from sklearn.cluster import KMeans
# Описываем модель
model = KMeans(n_clusters=3, n_jobs=-1, random_state=17)
 
# Проводим моделирование
model.fit(iris_df.data)
 
```

```python
# Предсказание на всем наборе данных
all_predictions = model.predict(iris_df.data)
#print(iris_df.data[0:5])
# Выводим предсказания
#count =len([x for x in all_predictions if x==0])
#print(count)
print(iris_df.target_names)
     
plt.scatter(x_axis, y_axis, c=all_predictions)
plt.show()
```

```python
# Импортируем библиотеки
from scipy.cluster.hierarchy import linkage, dendrogram
import pandas as pd
 
# Реализация иерархической кластеризации при помощи функции linkage
mergings = linkage(iris_df.data, method='complete')
#mergings = linkage(iris_df.data, method='average')
 
# Строим дендрограмму, указав параметры удобные для отображения
dendrogram(mergings,
            labels=iris_df.target,
           leaf_rotation=90,
           leaf_font_size=6,
           )
 
plt.show()
```

### TSNE

```python
# Импортируем библиотеки
from sklearn.manifold import TSNE
# Определяем модель и скорость обучения
model = TSNE(learning_rate=100)
 
# Обучаем модель
transformed = model.fit_transform(iris_df.data)

# Представляем результат в двумерных координатах
x_axis = transformed[:, 0]
y_axis = transformed[:, 1]
 
plt.scatter(x_axis, y_axis, c=iris_df.target)
plt.show()
```
