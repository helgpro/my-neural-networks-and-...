```python
x = [1,2,3,4,5]
y = [-1,1,2,1,-3]
err = [0.3,0.2,0.1,1,0.4]
plt.title("Simplest errorbars, 0.2 in x, 0.4 in y")
plt.errorbar(x,y, yerr=err,fmt='o',ecolor='black',capsize=4,ms=6)
plt.xlabel('Дата измерения')
```

```python
from datetime import datetime, date, time
x = [1,2,3,4,5,6]
y = [-0.10,0.10,0.20,-0.20,-0.291,0.291]
p = np.polyfit(x,y,1)
print('p=',p)
# y2=p[0]*x+p[1]*x
y2 = np.polyval(p, x)
print(y2)

err = [0.3,0.2,0.4,0.1,0.1,0.2]

fig = plt.figure(figsize=(10,5))
d = [date(2009, 11, 18),date(2009,11,19),date(2009,11,21),date(2009,11,22),date(2009,11,24),date(2009,11,25),]
plt.title("Simplest errorbars, 0.2 in x, 0.4 in y")
plt.errorbar(d,y, yerr=err,fmt='o',ecolor='black',capsize=4,ms=6)
plt.plot(d, y2, 'r')
plt.xlabel('Дата измерения')
```
