## Есть функция 10x^2
### ЕЕ производная 20x
#### xi = a*x^(-1) т.е.  xi = alfa*(x*20) где x от первой точки  = 0.001*(20*x) альфа гипер параметр

```python
#Сама функция но с ней понты в типах данных
def fw(x):
    return 10* x**2
    #   Тут  задаеться точка альфа от 0 до 1  и конечное значение куда должен прейти спуск eps
def gradspusk(alpha=0.01,eps=0.001):
    x_pred = 100  
    x = 50
    for _ in range(1000):# Задаем шаг щт 0 до 1000 round округляет 2-й знак сколько чисел посля запятой 
        #print("Шаг:", _, "\tX = ", round(x,5), "\tf(x)=", round(10*x**2,5))
        print("Шаг:", _, "\tX = ", round(x,4), "\tf(x)=", round(fw(x),5))
        #print("Шаг:", _, "\tX = ", x, "\tf(x)=", fw(x)) 
        #print(round(x,5), round(10*x**2,5))
        print(x ,'   ',x_pred, '    ',x-x_pred)
        if (x - x_pred)**2 < eps**2:# тут значения x - x второго это в квадрате чтоб от минусов избавиться и сравнев с eps
            break
        x_pred = x
        x = x - alpha * (20 * x_pred) # x - 0.001*(20*x-x_pred) короче херня
    
    return x
    
    
```

```python
x_min = gradspusk()
```

```python
#Сама функция но с ней понты в типах данных
def f(x):
    return 10* x[0]**2 + x[1]**2  
    #   Тут  задаеться точка альфа от 0 до 1  и конечное значение куда должен прейти спуск eps
def gradient_disend(alpha=0.01,eps=0.001):
    x_pred = np.array([100,100]) # создаст список [0 = 100,1 =100]
    x = np.array([50,50])
    for _ in range(100000):# Задаем шаг щт 0 до 1000 round округляет 2-й знак сколько чисел посля запятой 
        print("Шаг:", _, "\tX = ", np.round(x,3), "\tf(x)=", np.round(f(x),5))
        if np.sum((x-x_pred)**2) < eps**2:# тут значения x - x второго это в квадрате чтоб от минусов избавиться сумируеться и сравнев с eps
            break
        x_pred = x
        x = x_pred - alpha * np.array(20 * x_pred[0], 2 * x_pred[1]) # x - 0.001*(20*x-x_pred) короче херня
        print('x_pred0  ',x_pred[0], ' xpred[1]', x_pred[1])
    
    return x
```
